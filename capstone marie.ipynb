# Gesundheit in Deutschland aktuell

```python
df2.head()
```

```python
def lower_data(data=df2):
    data.columns=data.columns.str.lower()
    df_lower=data
    return df_lower
```

```python
df2_lower=lower_data(df2)
```

```python
df2_lower
```

### Variables that are relevant for us:

"IAarzt8C" - Inanspruchnahme: Psycholog:in

"PKPHQ8_k6 - Depressive Symptomatik (PHQ-8)"

```python
#"IAarzt8C" - Inanspruchnahme: Psycholog:in

variable_to_filter = 'IAarzt8C'
filtered_df = df2_lower[df2_lower['variable'] == variable_to_filter]
```

```python
filtered_df
```

```python
filtered_df2=filtered_df[filtered_df['altersgruppe'] != 'Gesamt']
```

```python
sns.histplot(data=filtered_df2, x='altersgruppe', hue='ja_nein', multiple='stack', shrink=0.8)

plt.title('Distribution of "Inanspruchnahme Psycholog:in"')
plt.xlabel('Altersgruppe')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(filtered_df2['altersgruppe'], bins=10)
plt.title('Distribution of "Inanspruchnahme Psycholog:in"')
plt.xlabel('altersgruppe')
plt.ylabel('ja/nein')
plt.show()
```
